## Funksjon for Ã¥ simulere fra en weibullfordeling

```python
# import numpy as np
import numpy as np
import matplotlib.pyplot as plt

def simX(alpha,beta,n):
    u = np.random.uniform(0,1,n)
    x = (-np.log(1-u)/alpha)**(1/beta)
    
    return x
```

## Simulere fra weibullfordeling og visualisere

```python
alpha = 1.2
beta = 1.5

n = 1000000

x = simX(alpha,beta,n)

plt.hist(x,100,density=True)
plt.xlabel("x")
plt.ylabel("f(x)")


plt.show()
```

```python
plt.hist(x,100,density=True)
plt.xlabel("x")
plt.ylabel("f(x)")

z = np.arange(0,5,0.01)
f = alpha * beta * z**(beta-1) * np.exp(-alpha*z**beta)
plt.plot(z,f)
plt.show()
```

## Approksimere $P(1 \leq X \leq 2)$

```python
mindreEnnTo = (x <= 2)
storreEnnEn = (x >= 1)
inniIntervallet = np.logical_and(mindreEnnTo,storreEnnEn)
sanns = np.sum(inniIntervallet)/n
print(sanns)

```

## Approksimere $E[X]$, $\mbox{Var}[X]$ og $\mbox{SD}[X]$

```python
xStrek = np.mean(x)
empiriskVarians = np.var(x,ddof=1)
empiriskStandardavvik = np.sqrt(empiriskVarians)

print('E[X]:', xStrek)
print('Var[X]:',empiriskVarians)
print('SD[X]:',empiriskStandardavvik)
```

## Tolkning av $E[X]$ og $\mbox{SD}[X]$

```python
plt.hist(x,100,density=True)
plt.xlabel("x")
plt.ylabel("f(x)")

z = np.arange(0,5,0.01)
f = alpha * beta * z**(beta-1) * np.exp(-alpha*z**beta)
plt.plot(z,f)

plt.plot([xStrek,xStrek],[0,0.9])
plt.plot([xStrek-empiriskStandardavvik,xStrek-empiriskStandardavvik],[0,0.9],'r')
plt.plot([xStrek+empiriskStandardavvik,xStrek+empiriskStandardavvik],[0,0.9],'r')
plt.show()
```
